Flask is a lightweight and popular web framework for Python. It is designed to be simple and easy to use, providing the necessary tools to build web applications. Flask is based on the Werkzeug WSGI toolkit and the Jinja2 template engine. It does not impose strict architectural patterns, allowing developers to choose how they structure their applications.

Advantages of Flask Framework:

1. Simplicity: Flask is straightforward and easy to understand, making it an excellent choice for small to medium-sized projects.

2. Flexibility: Flask gives developers the freedom to choose their libraries and extensions, allowing for a more customized development experience.

3. Minimalistic: Flask does not include unnecessary components, keeping the core small and efficient.

4. Built-in Development Server: Flask comes with a built-in development server, making it easy to test applications during development.

5. Extensible: Flask has a large ecosystem of extensions that can add various functionalities, such as database integration, authentication, and more.

6. Great Documentation: Flask has well-documented and accessible documentation, making it easy for developers to get started and find answers to their questions.

App routing in Flask refers to the process of defining the mapping between URLs and the functions that handle those URLs. In Flask, we use the @app.route() decorator to create routes.

The app.route() decorator is used to bind a URL to a Python function. When a user makes a request to a specific URL, Flask will invoke the function associated with that URL (route) and return the response.

We use app routes to create different endpoints for our web application. Each route represents a specific URL that a user can visit, and the associated function defines what should happen when that URL is accessed. App routing allows us to build a RESTful API or serve different HTML pages based on the URLs.

In Flask, the url_for() function is used for URL building. It generates a URL for a specified view function, taking into account the route defined using @app.route(). The url_for() function accepts the name of the view function as an argument and returns the corresponding URL.

Here's a simple Python code to demonstrate the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

if __name__ == '__main__':
    with app.test_request_context():
        print("URL for home():", url_for('home'))
        print("URL for about():", url_for('about'))
```
